````python
import pandas as pd
import numpy as np

# Cargar la lista de estudiantes desde la URL
path = '/home/marco/Documentos/docencia/groups_list/calculo_vectorial.csv'
estudiantes = pd.read_csv(path)
estudiantes['Parcial'] = estudiantes['Parcial'].astype(float) # hacer que todos los valores de la columna Parcial seen de tipo float

# Convertir la columna 'Nombre' a una lista de Python
lista_estudiantes = estudiantes['Nombre'].tolist()

# Función para llamar a un estudiante aleatoriamente y registrar su nota de seguimiento
def asistencia():
    if lista_estudiantes:
        estudiante_aleatorio = np.random.choice(lista_estudiantes)
        print(f"Llamando a {estudiante_aleatorio}...")
        asistencia = input(f"¿Asistió {estudiante_aleatorio}? (s/n): ").strip().lower()
        
        if asistencia == 's':
            # Obtener el índice del estudiante en el DataFrame
            indice = estudiantes[estudiantes['Nombre'] == estudiante_aleatorio].index[0]
            # Incrementar su valor en la columna 'Parcial'
            estudiantes.at[indice, 'Parcial'] += 0.1
            print(f"Actualizado: {estudiante_aleatorio} ahora tiene {estudiantes.at[indice, 'Parcial']} en la columna 'Parcial'.")
        
        lista_estudiantes.remove(estudiante_aleatorio)
    else:
        print("Se han llamado a todos los estudiantes del grupo completo.")

# Guardar los cambios en el archivo CSV
def guardar_cambios():
    estudiantes.to_csv(path, index=False)
    print("Cambios guardados en el archivo CSV.")

# Ejemplo de uso
while lista_estudiantes:
    asistencia()
    guardar_cambios()


````

#### Desarrollo de la Clase

### 1. Introducción a `pandas.sort_values()`



La función `sort_values()` se utiliza para ordenar los datos en un DataFrame o Series según los valores de una o más columnas. Esta función devuelve un nuevo DataFrame o Series ordenado.

**Sintaxis básica:**

```python
DataFrame.sort_values(by, axis=0, ascending=True, inplace=False, kind='quicksort', na_position='last')
```

- **by**: Nombre(s) de la(s) columna(s) por la(s) cual(es) se desea ordenar.
- **axis**: Define si se ordena por filas (`0` o `'index'`) o columnas (`1` o `'columns'`).
- **ascending**: Orden ascendente (`True`) o descendente (`False`).
- **inplace**: Si se debe modificar el DataFrame original (`True`) o devolver uno nuevo (`False`).
- **na_position**: Ubicación de los valores nulos ('first' para primeros, 'last' para últimos).

### 2. Ordenamiento por una Sola Columna

Comencemos creando un DataFrame de ejemplo:

````python

import pandas as pd

# Crear un DataFrame de ejemplo
data = {
    'Nombre': ['Ana', 'Carlos', 'Beatriz', 'David'],
    'Edad': [23, 24, 30, 15],
    'Salario': [50000, 64000, 70000, 52000]
}

df = pd.DataFrame(data)
df
````

**Ordenar por la columna 'Edad':**

````python
df_ordenado = df.sort_values(by='Edad', ascending = True )
df_ordenado
````

Esto ordenará el DataFrame en función de la columna 'Edad' en orden ascendente.

### 3. Ordenamiento por Múltiples Columnas

Podemos ordenar por más de una columna especificando una lista de nombres de columna en el parámetro `by`.

**Ordenar por 'Edad' y luego por 'Salario':**

````python

df_ordenado = df.sort_values(by=['Edad', 'Salario'])
df_ordenado
````

### 4. Orden Ascendente y Descendente

Podemos cambiar el orden de ascendente a descendente usando el parámetro `ascending`.

**Ordenar 'Salario' en orden descendente:**


````python

df_ordenado = df.sort_values(by='Salario', ascending=False)
df_ordenado

````


**Ordenar 'Edad' en ascendente y 'Salario' en descendente:**


````python
df_ordenado = df.sort_values(by=['Edad', 'Salario'], ascending=[True, False])
df_ordenado

````


### 5. Manejo de Valores Nulos

La posición de los valores nulos se puede controlar con el parámetro `na_position`.

**Ordenar 'Salario' colocando los nulos al inicio:**


````python

df.loc[2, 'Salario'] = None  # Introducir un valor nulo
df_ordenado = df.sort_values(by='Salario', na_position='first')
df_ordenado

````


### 6. Introducción a `pandas.sort_index()`

La función `sort_index()` se utiliza para ordenar un DataFrame o Series en función de sus etiquetas (índices) en lugar de los valores.

**Ejemplo:**


````python


df_ordenado_index = df.sort_index(ascending=True)
df_ordenado_index
````
