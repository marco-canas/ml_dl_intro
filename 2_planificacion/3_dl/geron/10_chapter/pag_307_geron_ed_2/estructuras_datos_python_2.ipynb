```python
import qrcode
from IPython.display import display

def generar_qr_notebook(url):
    """
    Genera un código QR para una URL y lo muestra en un cuaderno Jupyter.
    
    Parámetros:
    url (str): La URL para la cual se generará el código QR.
    
    Retorna:
    None (pero muestra el código QR en el notebook)
    """
    # Crear el objeto QR
    qr = qrcode.QRCode(
        version=1,
        error_correction=qrcode.constants.ERROR_CORRECT_L,
        box_size=10,
        border=4,
    )
    
    # Añadir la URL al código QR
    qr.add_data(url)
    qr.make(fit=True)
    
    # Crear la imagen del QR
    img = qr.make_image(fill_color="black", back_color="white")
    
    # Mostrar la imagen en el notebook
    display(img)

```


# **Cálculos Lógicos con Neuronas**  


Warren McCulloch y Walter Pitts propusieron un modelo muy simple de la neurona biológica, que luego se conoció como neurona artificial: tiene una o más entradas binarias (activado/desactivado) y una salida binaria. 

<img src = 'neurona_mcculloch_pitt_2.png' width = 400>

La neurona artificial simplemente activa su salida cuando más de un cierto número de sus entradas están activas. 

McCulloch y Pitts demostraron que incluso con un modelo tan simplificado es posible construir una red de neuronas artificiales que calcule cualquier proposición lógica que se desee.

* $\neg p$
* $p \lor q$
* $p \and q$
* $p \to q$
* $p \leftrightarrow q$
* y otras 

 Por ejemplo, construyamos algunas RNA (Redes Neuronales Artificiales) que realicen diversos cálculos lógicos (ver Figura 10-3), asumiendo que una neurona se activa cuando al menos dos de sus entradas están activas.


<img src = 'fig_10_3.png' width = 500>


**Figura 10-3. RNA realizando cálculos lógicos simples**  


- La primera red a la izquierda es simplemente la función identidad: si la neurona A está activada, entonces la neurona C también se activa (ya que recibe dos señales de entrada de la neurona A), pero si la neurona A está desactivada, la neurona C también lo estará. 

 
- La segunda red realiza un AND lógico: la neurona C se activa solo cuando tanto la neurona A como la B están activadas (una sola señal de entrada no es suficiente para activar la neurona C).  


- La tercera red realiza un OR lógico: la neurona C se activa si la neurona A o la B está activada (o ambas).  


- Finalmente, si suponemos que una conexión de entrada puede inhibir la actividad de la neurona (como ocurre en las neuronas biológicas), la cuarta red calcula una proposición lógica un poco más compleja: la neurona C se activa solo si la neurona A está activa y la neurona B está desactivada. Si la neurona A está siempre activa, se obtiene un NOT lógico: la neurona C está activa cuando la B está desactivada, y viceversa.  



Puedes imaginar fácilmente cómo estas redes pueden combinarse para calcular expresiones lógicas complejas (ver los ejercicios al final del capítulo 10 de Géron: Página 430 del PDF).




## **5. Proposición: $ A \land \neg B$ (A AND NOT B)**
Combina las neuronas AND y NOT para implementar la lógica $A \land \neg B$.


```python

A, B = 1, 1
not_B = neurona_not(B)
output = neurona_and(A, not_B)
print(output)  # Salida: 0 (porque B es 1)

```




# **6. Proposición: $\neg(A \lor B)$ (NOT (A OR B))**
Combina OR y NOT para implementar la negación de OR (NOR).


```python
A, B = 1, 0
or_output = neurona_or(A, B)
output = neurona_not(or_output)
print(output)  # Salida: 0 (porque A OR B es 1)
```


### **7. Proposición: $(A \lor B) \land \neg C$**
Combina OR, AND y NOT para implementar la lógica deseada.



```python
A, B, C = 1, 0, 1
or_output = neurona_or(A, B)
not_C = neurona_not(C)
output = neurona_and(or_output, not_C)
print(output)  # Salida: 0 (porque C es 1)


```



### **8. Proposición: XOR (A XOR B)**


Implementa XOR usando combinaciones de AND, OR y NOT:
$$ A \oplus B = (A \land \neg B) \lor (\neg A \land B)$


```python

A, B = 1, 0
not_A = neurona_not(A)
not_B = neurona_not(B)
and1 = neurona_and(A, not_B)
and2 = neurona_and(not_A, B)
output = neurona_or(and1, and2)
print(output)  # Salida: 1 (XOR de 1 y 0)

```


---

### **9. Proposición: Implicación $A \rightarrow B$ (NOT A OR B)**
Implementa la implicación lógica usando NOT y OR.



```python
A, B = 1, 0
not_A = neurona_not(A)
output = neurona_or(not_A, B)
print(output)  # Salida: 0 (porque A=1 y B=0)


```



### **10. Proposición: Equivalencia \( A \leftrightarrow B \)**
Implementa la equivalencia (XNOR) combinando XOR y NOT:
$$ A \leftrightarrow B = \neg (A \oplus B) $$



```python
A, B = 1, 1
xor_output = neurona_or(
    neurona_and(A, neurona_not(B)),
    neurona_and(neurona_not(A), B)
)
output = neurona_not(xor_output)
print(output)  # Salida: 1 (A y B son iguales)


```

#### **1. Práctica: Implementación de Neuronas Artificiales para Compuertas Lógicas**

  
**Objetivo**: Implementar neuronas artificiales que simulen compuertas lógicas (AND, OR, NOT, IDENTITY) usando el modelo de McCulloch-Pitts.


```python
import numpy as np
```

```python

class McCullochPittsNeuron:
    def __init__(self, weights, threshold):
        self.weights = weights
        self.threshold = threshold

    def activate(self, inputs):
        weighted_sum = np.dot(inputs, self.weights)
        return 1 if weighted_sum >= self.threshold else 0
```

```python

# Compuertas Lógicas
def test_logic_gate(gate_name, neuron, test_cases):
    print(f"\n{gate_name}:")
    for inputs in test_cases:
        output = neuron.activate(inputs)
        print(f"Input: {inputs} -> Output: {output}")
```

```python

# Configuraciones para cada compuerta
and_neuron = McCullochPittsNeuron(weights=[1, 1], threshold=2)
or_neuron = McCullochPittsNeuron(weights=[1, 1], threshold=1)
not_neuron = McCullochPittsNeuron(weights=[-1], threshold=0)
identity_neuron = McCullochPittsNeuron(weights=[2], threshold=1)
```

```python

# Casos de prueba
test_cases_and_or = [(0, 0), (0, 1), (1, 0), (1, 1)]
test_cases_not = [(0,), (1,)]
```

```python

# Evaluación
test_logic_gate("AND", and_neuron, test_cases_and_or)
test_logic_gate("OR", or_neuron, test_cases_and_or)
test_logic_gate("NOT", not_neuron, test_cases_not)
test_logic_gate("IDENTITY", identity_neuron, test_cases_not)


```



#### **2. Práctica de Visualización con Matplotlib y Plotly**  


**Objetivo**: Visualizar el comportamiento de las neuronas artificiales para diferentes entradas.


```python
and_neuron = McCullochPittsNeuron(weights=[1, 1], threshold=2)
or_neuron = McCullochPittsNeuron(weights=[1, 1], threshold = 1)
```

```python


import matplotlib.pyplot as plt
import plotly.graph_objects as go

# Datos para visualización (AND como ejemplo)
inputs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
outputs_and = np.array([and_neuron.activate(x) for x in inputs])

# Matplotlib
plt.figure(figsize=(8, 4))
plt.scatter(inputs[:, 0], inputs[:, 1], c=outputs_and, cmap='coolwarm', s=100)
plt.title("AND Gate Visualization (Matplotlib)")
plt.xlabel("Input A")
plt.ylabel("Input B")
plt.colorbar(label="Output (0 or 1)")
plt.grid(True)
plt.show()

# Plotly
fig = go.Figure(data=[go.Scatter3d(
    x=inputs[:, 0],
    y=inputs[:, 1],
    z=outputs_and,
    mode='markers',
    marker=dict(size=10, color=outputs_and, colorscale='Viridis')
)])
fig.update_layout(
    title="AND Gate Visualization (Plotly)",
    scene=dict(xaxis_title='Input A', yaxis_title='Input B', zaxis_title='Output')
)
fig.show()


```

## 1️. Introducción (lectura previa a clase)

En Python, las **estructuras de datos básicas** permiten almacenar y organizar información de manera eficiente.


En el mundo empresarial, son como las **herramientas de gestión** que facilitan el manejo de clientes, inventarios, ventas y reportes financieros.

* **Listas (`list`)**: colecciones ordenadas y modificables. Ejemplo: la lista de facturas emitidas en un mes.


* **Tuplas (`tuple`)**: colecciones ordenadas y que no se pueden cambiar. Ejemplo: los datos fijos de un producto (código, nombre, precio base).


* **Conjuntos (`set`)**: colecciones no ordenadas de elementos únicos. Ejemplo: el conjunto de clientes que compraron en una promoción (sin duplicados).


* **Diccionarios (`dict`)**: pares clave–valor, como un “mapa” de datos. Ejemplo: asociar cada cliente con el saldo que debe.


* **Cadenas (`str`)**: secuencias de caracteres, útiles para procesar texto. Ejemplo: analizar los mensajes de retroalimentación de clientes.

---



Así, lo que en programación son **estructuras de datos**, en administración se convierten en **recursos para organizar la información de la empresa** y tomar mejores decisiones.

El problema de tu script es que la regla para asignar la clase (`aprobacion`) depende de condiciones lógicas que no siempre generan **fronteras lineales** entre las clases para todos los pares de atributos.

Para garantizar que las clases sean **linealmente separables para cada par de atributos predictores**, podemos definir la etiqueta (`OtorgarTarjeta`) a partir de una **función lineal de los predictores** y luego aplicar un umbral (ejemplo: `>= 0 → clase 1, < 0 → clase 0`).



Aquí te dejo el script corregido:




```python
import numpy as np
import pandas as pd

# Fijamos semilla para reproducibilidad
np.random.seed(702)

# Número de muestras
n_samples = 20

# Variables simuladas
edad = np.random.randint(18, 70, n_samples)
ingresos_mensuales = np.random.randint(1_423_500, 10_000_000, n_samples)
nivel_endeudamiento = np.round(np.random.uniform(0, 1, n_samples), 2)
antiguedad_laboral = np.random.randint(0, 30, n_samples)

# Variables categóricas (pueden ser ignoradas para la separabilidad lineal)
historial_crediticio = np.random.choice(['Excelente', 'Regular', 'Malo'], n_samples, p=[1.0, 0.0, 0.0])
estado_civil = np.random.choice(['Soltero', 'Casado', 'Divorciado'], n_samples, p=[0.5, 0.4, 0.1])

# -------- Regla lineal para la clase -------- #
# Definimos una combinación lineal de los predictores
score = (
    0.03*edad +
    0.000001*ingresos_mensuales -
    0.8*nivel_endeudamiento +
    0.05*antiguedad_laboral
)

# Asignación de clase linealmente separable
aprobacion = (score > np.median(score)).astype(int)

# Crear DataFrame
df = pd.DataFrame({
    'Edad': edad,
    'IngresosMensuales': ingresos_mensuales,
    'HistorialCrediticio': historial_crediticio,
    'NivelEndeudamiento': nivel_endeudamiento,
    'AntiguedadLaboral': antiguedad_laboral,
    'EstadoCivil': estado_civil,
    'OtorgarTarjeta': aprobacion
})
```

```python

df.index = range(1, len(df)+1)
df.head(20)
```

### Cambios claves:

1. La clase `OtorgarTarjeta` ahora depende de una **función lineal de los predictores numéricos** (`Edad`, `IngresosMensuales`, `NivelEndeudamiento`, `AntiguedadLaboral`).
2. Uso de un **umbral en la mediana de `score`** para garantizar balance entre clases.
3. Esto asegura que cualquier par de predictores esté separado por una **frontera lineal** (recta en 2D).



# Función de python para visualizar pares de atributos predictores y variable objetivo  


Lo que necesitas es una **función de usuario** en Python que:

1. Reciba tu **DataFrame `df`** y dos nombres de columnas predictoras.
2. Genere el **diagrama de dispersión** de esos dos atributos.
3. Coloree los puntos según la variable objetivo (`OtorgarTarjeta`).

Aquí te dejo la función:




Esta es una función en Python que recorre **todos los pares de atributos numéricos predictores** de tu dataset y dibuja los diagramas de dispersión, coloreando por la variable objetivo `OtorgarTarjeta`. 

Además, dibuja la **frontera lineal de decisión** calculada a partir de una regresión logística entrenada en cada par:


```python
import itertools
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LogisticRegression

def graficar_pares_con_frontera(df, target='OtorgarTarjeta'):
    # Seleccionamos solo las columnas numéricas predictoras
    predictores = df.select_dtypes(include=[np.number]).drop(columns=[target]).columns
    
    pares = list(itertools.combinations(predictores, 2))
    
    for x_col, y_col in pares:
        X = df[[x_col, y_col]].values
        y = df[target].values
        
        # Entrenamos un modelo de regresión logística
        model = LogisticRegression()
        model.fit(X, y)
        
        # Rango de valores para la malla
        x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
        y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
        xx, yy = np.meshgrid(np.linspace(x_min, x_max, 200),
                             np.linspace(y_min, y_max, 200))
        
        # Predicciones en la malla
        Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
        Z = Z.reshape(xx.shape)
        
        # Gráfico
        plt.figure(figsize=(6, 5))
        plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.coolwarm)
        plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, edgecolor='k', s=70)
        plt.xlabel(x_col)
        plt.ylabel(y_col)
        plt.title(f"Frontera lineal entre {x_col} y {y_col}")
        plt.show()

```


### ¿Cómo usarla?

Después de generar tu dataset (`df`), simplemente ejecuta:


```python

graficar_pares_con_frontera(df, target='OtorgarTarjeta')

```


Esto te mostrará un **diagrama de dispersión con frontera lineal** para cada par de predictores (Edad vs IngresosMensuales, Edad vs NivelEndeudamiento, etc.).

```python
import matplotlib.pyplot as plt

def scatter_predictores(df, col_x, col_y, target='OtorgarTarjeta'):
    """
    Genera un diagrama de dispersión entre dos atributos predictivos,
    coloreando por la variable objetivo.

    Parámetros:
    -----------
    df : pandas.DataFrame
        Dataset con los atributos y la variable objetivo.
    col_x : str
        Nombre de la columna a ubicar en el eje X.
    col_y : str
        Nombre de la columna a ubicar en el eje Y.
    target : str (default='OtorgarTarjeta')
        Nombre de la variable objetivo para colorear.
    """
    plt.figure(figsize=(8,6))
    scatter = plt.scatter(
        df[col_x],
        df[col_y],
        c=df[target],
        cmap='bwr',   # azul = 0, rojo = 1
        alpha=0.7,
        edgecolors='k'
    )
    
    plt.xlabel(col_x, fontsize=12)
    plt.ylabel(col_y, fontsize=12)
    plt.title(f'Diagrama de dispersión: {col_x} vs {col_y}', fontsize=14)
    
    # Leyenda para clases
    handles, _ = scatter.legend_elements(prop="colors")
    labels = [f"{target}=0", f"{target}=1"]
    plt.legend(handles, labels, title="Clases")
    
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()
```

# **función de usuario en Python** que toma cualquier par de atributos predictores de tu dataset y grafica un **diagrama de dispersión coloreado por la variable objetivo `OtorgarTarjeta`**:



```python
import matplotlib.pyplot as plt

def scatter_predictores(df, x_col, y_col, target_col="OtorgarTarjeta"):
    """
    Genera un diagrama de dispersión para dos predictores del dataset,
    coloreando los puntos según la variable objetivo.

    Parámetros:
    -----------
    df : pandas.DataFrame
        Dataset con los predictores y la variable objetivo.
    x_col : str
        Nombre de la columna para el eje X.
    y_col : str
        Nombre de la columna para el eje Y.
    target_col : str
        Nombre de la columna objetivo (default: 'OtorgarTarjeta').

    Retorna:
    --------
    None (muestra el gráfico).
    """

    # Crear la figura
    plt.figure(figsize=(7, 5))

    # Graficar los puntos coloreados por la variable objetivo
    scatter = plt.scatter(
        df[x_col],
        df[y_col],
        c=df[target_col],
        cmap="coolwarm",  # Azul para 0, Rojo para 1
        s=100,            # Tamaño de puntos
        alpha=0.8,
        edgecolors="k"
    )

    # Títulos y etiquetas
    plt.title(f"Diagrama de dispersión: {x_col} vs {y_col}", fontsize=14)
    plt.xlabel(x_col, fontsize=12)
    plt.ylabel(y_col, fontsize=12)

    # Leyenda para la variable objetivo
    legend1 = plt.legend(*scatter.legend_elements(),
                         title=target_col,
                         loc="best")
    plt.gca().add_artist(legend1)

    plt.grid(True, linestyle="--", alpha=0.6)
    plt.show()

```
