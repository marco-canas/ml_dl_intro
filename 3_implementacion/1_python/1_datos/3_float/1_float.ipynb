```python
# Mapa conceptual de la secuencia metodológica del curso de machine learning
import networkx as nx
import matplotlib.pyplot as plt

# Crear un gráfico dirigido
G = nx.DiGraph()

plt.figure(figsize=(12, 12))

# Agregar nodos
nodos = ['Python', 'Estadística', 'Regresión', 'Clasificación']
G.add_nodes_from(nodos)

# Agregar bordes con etiquetas
bordes = [('Python', 'Estadística'), ('Estadística', 'Regresión'), \
          ('Regresión', 'Clasificación')]
G.add_edges_from(bordes)

# Dibujar el gráfico
pos = nx.spring_layout(G, seed = 3)
nx.draw_networkx_nodes(G, pos, node_size=2000, node_color='skyblue')
nx.draw_networkx_labels(G, pos, font_size=10, font_family='sans-serif')
nx.draw_networkx_edges(G, pos, width=2, arrows=True, edge_color='gray', arrowstyle='->')

# Agregar etiquetas a los bordes
etiquetas = {('Python', 'Estadística'): 'Estudio básico \n de estadísticas',
            ('Estadística', 'Regresión'): 'Aplicar \n regresión lineal',
            ('Regresión', 'Clasificación'): 'Comprender técnicas\n de clasificación',
            }

nx.draw_networkx_edge_labels(G, pos, edge_labels=etiquetas, font_color='red')

# Mostrar el gráfico
plt.axis('off')
plt.show()
```

```python
# Ubicacion de donde están en el curso
from graphviz import Digraph

# Crear un nuevo grafo dirigido
dot = Digraph()

# Añadir nodos y relaciones al grafo
dot.node('A', 'Python')
dot.node('B', 'Tipos de editores de código')
dot.node('C', 'Tipos de datos')
dot.node('D', 'Datos tipo str')
dot.node('E', 'Datos tipo int')
dot.node('F', 'Datos tipo float')
dot.node('G', '')
dot.node('H', '')
dot.node('I', '')
dot.node('J', '')

dot.edges(['AB'])
dot.edges(['AC'])
dot.edges(['CD'])
dot.edges(['DE'])
dot.edges(['EF'])
dot.edges(['FG'])
dot.edges(['GH'])
dot.edges(['HI'])
dot.edges(['IJ'])

# Guardar el grafo como un archivo de imagen
dot.render(filename='ecuaciones', format='png', cleanup=True, \
           directory='/home/marco/Documentos/docencia/calculo/images/0_reales/')

dot.view()
```
