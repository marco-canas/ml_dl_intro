### **Metodología de Enseñanza: Aprendizaje Basado en Problemas y Proyectos (ABPP)**



El **Aprendizaje Basado en Problemas y Proyectos (ABPP)** es una metodología educativa que integra los enfoques del **Aprendizaje Basado en Problemas (ABP)** y el **Aprendizaje Basado en Proyectos (PBL, por sus siglas en inglés)**. Es especialmente útil para la formación de futuros docentes, ingenieros y administradores, ya que combina el desarrollo de habilidades prácticas con el pensamiento crítico y reflexivo. A continuación, se presenta una descripción adaptada a la enseñanza de las matemáticas en educación infantil, Licenciatura en Matemáticas, Ingeniería Agropecuaria y Administración de Empresas:

---



### **1. Fundamentos del ABPP**
El ABPP se basa en:
- **Resolución de Problemas Reales:** Los estudiantes analizan y resuelven problemas auténticos relacionados con su área de formación, fomentando la aplicación práctica del conocimiento.
- **Proyectos como Herramientas Integradoras:** Los proyectos permiten desarrollar habilidades interdisciplinarias y ofrecer soluciones concretas a los problemas planteados.
- **Aprendizaje Activo:** Los estudiantes son responsables de su aprendizaje y participan activamente en el proceso de investigación, diseño y solución.

---

#### Definición de variables binarias y continuas



Una **variable binaria** es aquella que solo puede tomar dos valores distintos, usualmente representados como 0 y 1. 
Ejemplo: Género (0 = Masculino, 1 = Femenino), aprobación de un examen (0 = No aprobado, 1 = Aprobado).



Una **variable continua** puede tomar un rango infinito de valores dentro de un intervalo dado.
Ejemplo: Peso en kilogramos, altura en metros, temperatura en grados Celsius.




# Concepto de correlación en presencia de variables binarias



La correlación mide la relación entre dos variables. En el caso de una variable binaria $X$ y una continua $Y$, se puede utilizar la **correlación punto biserial** $r_{pb}$, definida como:


$$
r_{pb} = \frac{\bar{Y}_1 - \bar{Y}_0}{s_Y} \sqrt{\frac{p(1-p)}{n}}
$$
Donde:
- $\bar{Y}_1$ y $\bar{Y}_0$ son las medias de $Y$ cuando $X=1$ y $X=0$.
- $s_Y$ es la desviación estándar de $Y$.
- $p$ es la proporción de casos donde $X=1$.
- $n$ es el número total de observaciones.


El coeficiente de correlación biserial puntual $r_{pb}$ mide la relación lineal entre una variable dicotómica $X$ y una variable continua $Y$. A partir de su definición:

$$
r_{pb} = \frac{\bar{Y}_1 - \bar{Y}_0}{s_Y} \sqrt{\frac{p(1-p)}{n}}
$$

se pueden hacer las siguientes interpretaciones:



1. **Diferencia estandarizada de medias**:  
   - $r_{pb}$ es proporcional a la diferencia entre los promedios de $Y$ en los grupos $X = 1$ y $X = 0$, es decir, cuanto mayor sea $|\bar{Y}_1 - \bar{Y}_0|$, mayor será $r_{pb}$.
   - Si $\bar{Y}_1 > \bar{Y}_0$, $r_{pb}$ es positivo, indicando que valores más altos de $X$ tienden a estar asociados con valores más altos de $Y$.
   - Si $\bar{Y}_1 < \bar{Y}_0$, $r_{pb}$ es negativo, indicando que $X = 1$ está asociado con valores menores de $Y$.



2. **Normalización por la desviación estándar de $Y$**:  
   - La diferencia de medias se divide por $s_{Y}$, lo que significa que $r_{pb}$ mide la diferencia en términos de desviaciones estándar, haciéndolo independiente de la escala de $Y$.

3. **Efecto de la proporción $p$**:  
   - El término $\sqrt{\frac{p(1-p)}{n}}$ ajusta la correlación según la proporción de observaciones en cada categoría de $X$.
   - Si $p \approx 0.5$, la varianza de $X$ es máxima y $r_{pb}$ es más representativo de la relación entre $X$ y $Y$.
   - Si $p$ es muy cercano a 0 o 1, el coeficiente se reduce porque la variabilidad en $X$ es baja.



4. **Escala de $r_{pb}$**:  
   - Como en cualquier coeficiente de correlación, $r_{pb}$ varía entre $-1$ y $1$, con valores extremos indicando una fuerte asociación entre $X$ y $Y$, y valores cercanos a 0 indicando una relación débil o inexistente.

En resumen, $r_{pb}$ cuantifica la fuerza y dirección de la relación entre $X$ y $Y$, basándose en la diferencia de medias, la dispersión de $Y$, y la distribución de $X$.

#### Cálculo de la correlación punto biserial y coeficiente de biserialidad en Python

```python
import numpy as np
import scipy.stats as stats

# Generar datos simulados
#np.random.seed(42)
n = 10
X = np.random.choice([0, 1], size=n, p = [0.01, 0.99])  # Variable binaria
Y = np.random.normal(loc = 5 + 2*X, scale = 1, size=n)  # Variable continua con media diferente según X

# Calcular correlación punto biserial
r_pb, p_value = stats.pointbiserialr(X, Y)
print(f"Correlación punto biserial: {r_pb:.3f}, p-valor: {p_value:.3f}")

```


# Visualización de datos segmentados por la variable binaria


```python
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import numpy as np
import ipywidgets as widgets
from IPython.display import display

def update_plot(y0, y1, sy, p):
    np.random.seed(42)
    n = 100  # Número total de observaciones
    X = np.random.choice([0, 1], size=n, p=[1 - p, p])  # Variable binaria con proporción p
    Y = np.random.normal(loc=y0 + (y1 - y0) * X, scale=sy, size=n)  # Variable continua
    
    # Calcular correlación punto biserial
    r_pb, _ = stats.pointbiserialr(X, Y)
    
    # Crear la figura
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=X, y=Y, hue=X, palette=['blue', 'red'], legend=False)
    plt.xlabel("Variable Binaria (X)")
    plt.ylabel("Variable Continua (Y)")
    plt.title(f"Distribución de Y según X\n Correlación biserial puntual: r = {r_pb:.3f}")
    plt.grid(alpha=0.5)
    plt.show()

# Crear deslizadores
slider_y0 = widgets.FloatSlider(min=0, max=10, step=0.1, value=5, description=r'$\bar{Y}_0$')
slider_y1 = widgets.FloatSlider(min=0, max=10, step=0.1, value=7, description=r'$\bar{Y}_1$')
slider_sy = widgets.FloatSlider(min=0.1, max=5, step=0.1, value=1, description='$s_Y$')
slider_p = widgets.FloatSlider(min=0.1, max=0.9, step=0.05, value=0.5, description='$p$')

# Crear la interacción
display(widgets.interactive(update_plot, y0=slider_y0, y1=slider_y1, sy=slider_sy, p=slider_p))
```

#### Animaciones y widgets interactivos para explorar correlaciones


```python

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import scipy.stats as stats
from IPython.display import HTML

# Generar datos simulados
np.random.seed(42)
n = 100
X = np.random.choice([0, 1], size=n, p = [0.1, 0.9])  # Variable binaria
Y = np.random.normal(loc = 2*X + 3, scale = 0.1, size=n)  # Variable continua con media diferente según X

# Crear la figura y los ejes
fig, ax = plt.subplots()
sc = ax.scatter(X, Y, c=X, cmap='coolwarm', alpha=0.6)
ax.set_xlabel("Variable Binaria (X)")
ax.set_ylabel("Variable Continua (Y)")
ax.set_title("Exploración de correlación punto biserial")
ax.grid(alpha=0.5)

# Agregar texto para mostrar la correlación y los conteos
correlation_text = ax.text(0.05, 0.95, '', transform=ax.transAxes, fontsize=12, verticalalignment='top')
count_text = ax.text(0.05, 0.85, '', transform=ax.transAxes, fontsize=12, verticalalignment='top')

def update(frame):
    np.random.shuffle(Y)  # Reorganizar Y para mostrar variabilidad
    sc.set_offsets(np.c_[X, Y])
    
    # Calcular correlación punto biserial
    r_pb, _ = stats.pointbiserialr(X, Y)
    correlation_text.set_text(f'Correlación punto biserial: {r_pb:.3f}')
    
    # Calcular el conteo de elementos por valor de X
    count_0 = np.sum(X == 0)
    count_1 = np.sum(X == 1)
    count_text.set_text(f'X=0: {count_0}, X=1: {count_1}')
    
    return sc, correlation_text, count_text

ani = animation.FuncAnimation(fig, update, frames=20, interval=500, blit=True)
HTML(ani.to_jshtml())

```

# Aquí tienes algunas interpretaciones de la animación:

1. **Variabilidad en la correlación**: La animación muestra cómo la dispersión de la variable continua $Y$ afecta la relación con la variable binaria $X$. Al reorganizar $Y$, se pueden observar fluctuaciones en la correlación punto biserial.

2. **Dependencia entre las variables**: Si los valores de $Y$ permanecen bien separados entre los grupos $X=0$ y $X=1$, la correlación es fuerte. Si $Y$ se mezcla entre ambos grupos, la correlación es débil o inexistente.

3. **Impacto del reordenamiento aleatorio**: Como $Y$ se reorganiza en cada cuadro de la animación, el patrón de dispersión cambia, mostrando cómo la estructura de los datos afecta la correlación.

4. **Variación del coeficiente punto biserial**: La animación ayuda a visualizar cómo un cambio en la distribución de $Y$ dentro de cada categoría de $X$ influye en la correlación observada.

5. **Patrón de agrupamiento**: En algunos cuadros de la animación, los puntos pueden agruparse más en torno a ciertos valores de $Y$, reflejando una posible relación subyacente más fuerte o más débil entre las variables.

# Cómo establecer correlación lineal entre una variable nnumérica continua $y$ y una categórica binaria $x$. 

Establecer la correlación entre una variable numérica continua, como el precio de vivienda ($y$), y una variable categórica binaria codificada de forma booleana ($x$), implica analizar cómo la variable continua varía en función de los dos valores posibles de la variable categórica (por ejemplo, $x = 0$ o $x = 1$).



Aquí hay algunos pasos detallados para realizar este análisis:

---



### 1. **Exploración Descriptiva**
   - Divida los datos en dos grupos según los valores de la variable categórica $(x = 0$ y $x = 1)$.
   - Calcule estadísticas descriptivas para cada grupo:
     - Media, mediana, desviación estándar, y rango del precio de vivienda para cada grupo.
   - Visualice la relación con gráficos:
     - Boxplots: Permiten comparar la distribución del precio de vivienda entre las dos categorías.
     - Violin plots: Ayudan a visualizar la densidad de los valores en ambas categorías.

---



### 2. **Prueba de Diferencias de Medias**
   - Realice una **prueba t de Student** (si los datos son aproximadamente normales) o una prueba de **Wilcoxon-Mann-Whitney** (si los datos no son normales) para evaluar si la diferencia de medias del precio entre las dos categorías es estadísticamente significativa.
     - Hipótesis nula $(H_0)$: No hay diferencia en el precio promedio entre las dos categorías.
     - Hipótesis alternativa $(H_1)$: Hay una diferencia en el precio promedio entre las dos categorías.

---



### 3. **Cálculo de una Métrica de Correlación**
   Como la variable categórica es binaria, una métrica común es calcular el **coeficiente de correlación de punto biserial**:
   $$
   r_{pb} = \frac{\overline{y_1} - \overline{y_0}}{s_y} \cdot \sqrt{\frac{n_1 \cdot n_0}{n^2}}
   $$
   Donde:
   - $\overline{y_1}$ y $\overline{y_0}$ son las medias de $y$ para $x = 1$ y $x = 0$, respectivamente.
   - $s_y$ es la desviación estándar de $y$ en toda la muestra.
   - $n_1$ y $n_0$ son los tamaños de los grupos para $x = 1$ y $x = 0$, respectivamente.
   - $n$ es el tamaño total de la muestra.

---



### 4. **Modelos Estadísticos**
   - Ajuste un **modelo lineal** simple:
     $$
     y = \beta_0 + \beta_1 x + \epsilon
     $$
     Donde $x$ es la variable categórica. El coeficiente $\beta_1$ indica la diferencia promedio en el precio de vivienda entre las dos categorías ($x = 0$ y $x = 1$).
   - Evalúe la significancia de $\beta_1$ para determinar si la variable categórica tiene un efecto relevante.

---



### 5. **Validación y Visualización**


   - Verifique los supuestos del análisis (por ejemplo, normalidad de los residuos para pruebas paramétricas).
   - Realice visualizaciones complementarias, como diagramas de dispersión (scatter plots) donde los puntos se coloreen por la variable categórica.

---

### Ejemplo en Python
Aquí tienes un ejemplo práctico para realizar este análisis usando `pandas` y `scipy`:


```python
import pandas as pd
import numpy as np
from scipy.stats import pointbiserialr, ttest_ind
import seaborn as sns
import matplotlib.pyplot as plt

# Datos ficticios
data = pd.DataFrame({
    'price': np.random.normal(loc = 200_000, scale = 50_000, size = 100),
    'has_garden': np.random.choice([0, 1], 100)  # Variable categórica binaria
})

# Estadísticas descriptivas
group_stats = data.groupby('has_garden')['price'].describe()
print(group_stats)

# Visualización
sns.boxplot(x='has_garden', y='price', data=data)
plt.title("Distribución del precio por tener jardín o no")
plt.grid(alpha = 0.5)
plt.show()

# Prueba t
group0 = data[data['has_garden'] == 0]['price']
group1 = data[data['has_garden'] == 1]['price']
t_stat, p_value = ttest_ind(group0, group1)
print(f"T-Stat: {t_stat}, P-Value: {p_value}")

# Correlación punto biserial
r_pb, p_value_rpb = pointbiserialr(data['has_garden'], data['price'])
print(f"Correlación punto biserial: {r_pb}, P-Value: {p_value_rpb}")

```
