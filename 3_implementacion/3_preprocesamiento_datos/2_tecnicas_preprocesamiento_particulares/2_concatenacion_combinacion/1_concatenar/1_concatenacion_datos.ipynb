# La función `pandas.concat()` 

### Clase: Uso de la función `pandas.concat()` en Ciencia de Datos con Python

**Objetivo de aprendizaje**:  
Al finalizar esta clase, los estudiantes serán capaces de utilizar la función `pandas.concat()` para concatenar o combinar diferentes objetos `DataFrame` y `Series` en Python, una habilidad esencial para la manipulación y análisis de datos en ciencia de datos.

---

#### 1. **Introducción a `pandas.concat()`**  

* La función `pandas.concat()` es una herramienta del módulo `pandas` que permite combinar objetos de datos (como `DataFrame` y `Series`) a lo largo de un eje (filas o columnas). 
* Es útil cuando necesitamos unir varias tablas que comparten o no un esquema similar.

---

#### 2. **Sintaxis de `pandas.concat()`**
```python
pandas.concat(objs, axis=0, join='outer', ignore_index=False, keys=None, levels=None, names=None, verify_integrity=False, sort=False, copy=True)
```

# Parámetros importantes:  

- **`objs`**: Lista o tupla de objetos a concatenar (puede ser `DataFrame` o `Series`).
  
- **`axis`**: Define el eje de concatenación: `0` para filas (por defecto, uno debajo del otro) y `1` para columnas(concatena uno al dado del otro).
  
- **`join`**: Tipo de unión; `outer` (por defecto) para unión externa (mantiene todos los índices) y `inner` para unión interna (solo los índices comunes).
  
- **`ignore_index`**: Si es `True`, ignora los índices originales y asigna nuevos índices.
  
- **`keys`**: Etiquetas para los diferentes objetos concatenados.
  
---

#### 3. **Ejemplo básico de concatenación de `DataFrames`**

Imaginemos que tenemos dos `DataFrames` con información de ventas de dos sucursales de una tienda:

````python

import pandas as pd

# Datos de la sucursal A
df_a = pd.DataFrame({
    'Producto': ['A', 'B', 'C'],
    'Ventas': [10, 20, 30]
})

# Datos de la sucursal B
df_b = pd.DataFrame({
    'Producto': ['D', 'E', 'F'],
    'Ventas': [40, 50, 60]
})

# Concatenar por filas (axis=0)
df_concat = pd.concat([df_a, df_b], axis=0)
df_concat
````


En este ejemplo, los dos `DataFrames` se concatenan uno debajo del otro (a lo largo de las filas), creando un nuevo `DataFrame`.

---

#### 4. **Concatenación por columnas (`axis=1`)**

Podemos unir los `DataFrames` por columnas. Veamos un ejemplo donde ambos `DataFrames` tienen diferentes columnas:

````python
df_c = pd.DataFrame({
    'Producto': ['A', 'B', 'C'],
    'Descuento': [5, 10, 15]
})

# Concatenar por columnas (axis=1)
df_concat_col = pd.concat([df_a, df_c], axis=1)
df_concat_col
````

---

#### 5. **Uso de `ignore_index`**

Si queremos ignorar los índices originales de los `DataFrames` y crear un nuevo índice secuencial:

````python

df_concat_index = pd.concat([df_a, df_b], ignore_index=True)
df_concat_index
````

---

#### 6. **Concatenación con `join`**

Si los `DataFrames` tienen diferentes columnas, el parámetro `join` define cómo queremos realizar la concatenación. Si usamos `join='inner'`, solo se unirán las columnas que ambos `DataFrames` comparten:

````python

df_d = pd.DataFrame({
    'Producto': ['A', 'B', 'C'],
    'Stock': [100, 150, 200]
})

# Concatenar con join 'inner'
df_concat_inner = pd.concat([df_a, df_d], join='inner', axis=1)
df_concat_inner
````

#### 7. **Asignación de claves con `keys`**

Podemos agregar una clave a cada objeto concatenado para identificar de dónde proviene cada fila:

````python

df_concat_keys = pd.concat([df_a, df_b], keys=['Sucursal A', 'Sucursal B'])
df_concat_keys
````

---

#### 8. **Ejercicio práctico**

2. Usa `pandas.concat()` para unir los datos de ambos años en un único `DataFrame`, y luego filtra las ventas del primer trimestre.
3. Experimenta con los parámetros `axis`, `join`, y `ignore_index` para observar cómo cambia el resultado.

---

### Conclusión  

* La función `pandas.concat()` es una herramienta flexible para combinar datos en proyectos de ciencia de datos. 
* Puede manejar datos de diferentes fuentes y permite adaptarse a distintas estructuras.
* La práctica con este método es clave para manejar conjuntos de datos más complejos en el análisis de datos.
