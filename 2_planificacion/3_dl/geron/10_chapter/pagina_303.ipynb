### 2. Importar las libreras necesarias


````python

import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

````


---

### 3. Cargar un conjunto de datos sencillo

Usaremos el dataset **MNIST** (dgitos manuscritos) para que los estudiantes visualicen cmo una red neuronal clasifica imgenes.


````python
# Cargar dataset de dgitos manuscritos
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
````

````python

# Normalizar los valores (de 0-255 a 0-1)
x_train = x_train.astype("float32") / 255
x_test = x_test.astype("float32") / 255
````

````python

# Aplanar imgenes de 28x28 a 784
x_train = x_train.reshape(-1, 784)
x_test = x_test.reshape(-1, 784)

print("Dimensiones de entrenamiento:", x_train.shape, y_train.shape)

````


---

### 4. Construir la red neuronal



Creamos un **Perceptrn Multicapa (MLP)** con dos capas ocultas.


````python

# Definir el modelo
model = keras.Sequential([
    layers.Dense(128, activation="relu", input_shape=(784,)),
    layers.Dense(64, activation="relu"),
    layers.Dense(10, activation="softmax") # 10 clases (dgitos del 0 al 9)
])
````

````python

# Compilar el modelo
model.compile(optimizer="adam",
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"])

````


---

### 5. Entrenar el modelo



````python

history = model.fit(x_train, y_train, 
                    validation_data=(x_test, y_test),
                    epochs=5, batch_size=32)


````

````python


### 6. Evaluar el modelo
````

````python


test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)
print("\nPrecisin en datos de prueba:", test_acc)

````

````python



### 7. Probar predicciones
````



predicciones = model.predict(x_test[:5])
