El enfoque STEAMS en educación preescolar, primaria y secundaria es una extensión del enfoque STEM (Ciencia, Tecnología, Ingeniería y Matemáticas) y su variante STEAM, que incorpora el Arte. El acrónimo STEAMS incluye:

1. **Science (Ciencia):** Fomenta la curiosidad por el mundo natural y el desarrollo de habilidades científicas, como la observación, experimentación y análisis.

2. **Technology (Tecnología):** Integra herramientas tecnológicas para potenciar el aprendizaje y la resolución de problemas, promoviendo la alfabetización digital desde edades tempranas.

6. **Social Sciences (Ciencias Sociales):** Incorpora la comprensión del entorno social, político, cultural y económico, promoviendo el desarrollo de la conciencia crítica sobre el mundo y su historia, además de habilidades de empatía y ciudadanía.

# Cómo generar una mapa de calor a partir de un mapa político 

Para generar un mapa de calor a partir de un mapa político por barrios en formato `.jpg` y vincularlo con una variable numérica utilizando Python, puedes seguir estos pasos básicos:

### Herramientas necesarias:
1. **OpenCV**: para cargar y manipular la imagen del mapa político.
2. **NumPy**: para trabajar con los datos y las matrices.
3. **Matplotlib** o **Plotly**: para crear y visualizar el mapa de calor.
4. **Pandas**: para gestionar los datos numéricos asociados a los barrios.

### Requerimientos:
```bash
pip install opencv-python numpy pandas matplotlib
```

### Algoritmo paso a paso:

1. **Cargar el mapa político**: Lee la imagen `.jpg` del mapa usando OpenCV.
2. **Identificar los barrios**: Debes definir una forma de identificar los distintos barrios del mapa. Esto puede implicar seleccionar manualmente las áreas correspondientes a cada barrio o utilizar técnicas como la segmentación de imágenes para separar los barrios.
3. **Asociar los valores numéricos a los barrios**: Usa un diccionario o un DataFrame de `pandas` que asocie el nombre del barrio con el valor numérico que deseas usar para el mapa de calor.
4. **Colorear el mapa de acuerdo con los valores**: Para cada barrio identificado, superpón el color en función del valor de la variable numérica, donde el color irá de verde (valor bajo) a rojo (valor alto).
5. **Mostrar el mapa de calor**: Utiliza `Matplotlib` para visualizar el resultado.

### Código de ejemplo:


````python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 1. Cargar el mapa político en formato .jpg
mapa = cv2.imread('mapa_politico.jpg')

# Convertir la imagen a RGB (OpenCV carga en formato BGR por defecto)
mapa_rgb = cv2.cvtColor(mapa, cv2.COLOR_BGR2RGB)

# 2. Crear un diccionario con los valores numéricos por barrio
# Puedes ajustar estos valores según tus datos reales.
datos_barrios = {
    'Barrio1': 20,
    'Barrio2': 50,
    'Barrio3': 70,
    'Barrio4': 90
}

# 3. Definir las coordenadas de los barrios en la imagen (esto puede requerir ajustes manuales)
# Estas son solo posiciones aproximadas. En la realidad, deberías identificar las posiciones exactas.
coordenadas_barrios = {
    'Barrio1': (100, 150),
    'Barrio2': (200, 250),
    'Barrio3': (300, 350),
    'Barrio4': (400, 450)
}

# 4. Generar un mapa de colores basado en los valores numéricos
def valor_a_color(valor, min_valor, max_valor):
    # Normalizar el valor a [0, 1]
    norm_valor = (valor - min_valor) / (max_valor - min_valor)
    
    # Generar un color entre verde (valor bajo) y rojo (valor alto)
    return (1 - norm_valor, norm_valor, 0)  # RGB para un gradiente verde-rojo

# 5. Determinar los valores mínimo y máximo para normalizar los colores
min_valor = min(datos_barrios.values())
max_valor = max(datos_barrios.values())

# 6. Crear una copia del mapa original para colorear
mapa_coloreado = np.copy(mapa_rgb)

# 7. Colorear los barrios en el mapa
for barrio, valor in datos_barrios.items():
    # Obtener la coordenada del barrio
    coord = coordenadas_barrios[barrio]
    
    # Obtener el color correspondiente
    color = valor_a_color(valor, min_valor, max_valor)
    
    # Convertir el color de [0, 1] a [0, 255] para la imagen
    color_rgb = tuple(int(c * 255) for c in color)
    
    # Colorear un círculo en la coordenada del barrio (se puede ajustar para más precisión)
    cv2.circle(mapa_coloreado, coord, 30, color_rgb, -1)

# 8. Mostrar el mapa de calor
plt.imshow(mapa_coloreado)
plt.axis('off')  # Ocultar los ejes
plt.show()

````


### Explicación del código:

1. **Lectura de la imagen**: El mapa se carga usando `cv2.imread`, y se convierte a formato RGB para poder usarlo con `Matplotlib`.
2. **Diccionario de barrios y valores**: Se asocia cada barrio con un valor numérico.
3. **Coordenadas de los barrios**: De manera simplificada, se usan coordenadas manuales para cada barrio. En un caso real, podrías obtener estas áreas de forma más automática, pero este ejemplo ilustra cómo asociar los valores a ubicaciones en el mapa.
4. **Coloreado**: Para cada barrio, se calcula un color que varía de verde a rojo, y luego se dibuja un círculo coloreado en la ubicación correspondiente al barrio en el mapa.
5. **Visualización**: Se utiliza `Matplotlib` para mostrar el mapa coloreado.

### Notas:
- La identificación automática de los barrios puede ser más compleja y puede requerir técnicas avanzadas como segmentación de imágenes o procesamiento manual con una herramienta externa.
- El método para generar los colores usa una función que normaliza los valores entre el mínimo y el máximo y genera un color entre verde y rojo en formato RGB.
- En lugar de círculos, podrías trabajar directamente con máscaras para colorear áreas específicas del mapa.

````python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize
import matplotlib.cm as cm

# Cargar mapa de Colombia (GeoJSON o archivo shape)
colombia_map = gpd.read_file(r"C:\Users\marco\Documentos\investigacion\didactica_ciencia_datos\1_curso_machine_learning\4_visualizacion\1_mapas_geograficos_de_calor\col_admbnda_adm0_mgn_itos_20200416.shp")





# Datos de capitales (ubicación, población y número de universidades)
# Puedes ajustar los datos según tu fuente
data = {
    'Capital': ['Bogotá', 'Medellín', 'Cali', 'Barranquilla', 'Cartagena', 'Bucaramanga', 'Pereira'],
    'Lat': [4.711, 6.251, 3.451, 10.968, 10.391, 7.119, 4.814],
    'Lon': [-74.072, -75.563, -76.532, -74.806, -75.479, -73.125, -75.694],
    'Poblacion': [7743955, 2529403, 2228000, 1202000, 914552, 528575, 490335],
    'Universidades': [45, 30, 25, 15, 12, 10, 8]
}

capitals = pd.DataFrame(data)

# Crear GeoDataFrame para las capitales
geometry = gpd.points_from_xy(capitals['Lon'], capitals['Lat'])
capitals_gdf = gpd.GeoDataFrame(capitals, geometry=geometry, crs="EPSG:4326")

# Crear la visualización
fig, ax = plt.subplots(1, 1, figsize=(10, 12))

# Dibujar mapa base de Colombia
colombia_map.plot(ax=ax, color='lightgrey', edgecolor='black')

# Normalizar colores según el número de universidades
norm = Normalize(vmin=capitals['Universidades'].min(), vmax=capitals['Universidades'].max())
cmap = cm.get_cmap('viridis')

# Añadir puntos de capitales al mapa
scatter = ax.scatter(
    capitals['Lon'],
    capitals['Lat'],
    s=capitals['Poblacion'] / 10000,  # Escalar tamaño por población
    c=capitals['Universidades'],      # Color por número de universidades
    cmap=cmap,
    alpha=0.7,
    edgecolor='k',
    label='Capitales'
)

# Añadir colorbar para representar número de universidades
cbar = fig.colorbar(cm.ScalarMappable(norm=norm, cmap=cmap), ax=ax)
cbar.set_label('Número de Universidades', fontsize=12)

# Etiquetas de las capitales
for idx, row in capitals.iterrows():
    ax.text(
        row['Lon'], row['Lat'], row['Capital'],
        fontsize=10, ha='right', va='bottom', color='black'
    )

# Configurar título y leyenda
ax.set_title('Mapa Político de Colombia con Información de Capitales', fontsize=16)
ax.set_xlabel('Longitud')
ax.set_ylabel('Latitud')
plt.legend(loc='lower right', title='Capitales')
plt.show()
````

### **Actividad de Clase: Expresar Ideas, Dudas y Emociones de Manera Respetuosa y Asertiva**



**Objetivo:**  
Desarrollar habilidades de comunicación efectiva en los estudiantes para que puedan expresar sus ideas, dudas y emociones de forma respetuosa y asertiva.

**Duración:**  
15-20 minutos

---



### **1. Introducción (5 minutos)**  
- **Tema:** Introducir a los estudiantes al concepto de la comunicación asertiva, explicando cómo expresar pensamientos y emociones de manera clara y respetuosa sin agredir o ser pasivo.
- **Dinámica:** Mostrar dos ejemplos contrastantes: uno de comunicación pasiva/agresiva y otro de comunicación asertiva. Preguntar a los estudiantes cuál creen que es más efectivo y por qué.

**Materiales:**  
- Pizarra o diapositivas con ejemplos de frases pasivas, agresivas y asertivas.
  
**Ejemplo:**  
- **Pasivo:** “Bueno… no sé si es una buena idea, pero… tal vez podrías intentarlo.”
- **Agresivo:** “¡Tu idea es terrible, no va a funcionar!”
- **Asertivo:** “Creo que podríamos mejorar esa idea si consideramos otro enfoque.”

---

### **2. Ejercicio en Parejas (10 minutos)**  
- **Instrucciones:** Dividir a los estudiantes en parejas. Darles diferentes situaciones cotidianas en las que deben comunicarse (pedir ayuda, expresar desacuerdo, compartir una idea, etc.). Cada pareja debe crear una breve conversación donde practiquen la comunicación asertiva.
  
**Situaciones propuestas:**
  - Expresar desacuerdo con un compañero sobre una tarea de grupo.
  - Pedirle ayuda al profesor sobre un concepto difícil.
  - Expresar frustración por un malentendido en clase.

- **Reglas:** Cada estudiante debe asegurarse de usar lenguaje respetuoso y asertivo para expresar su posición y emociones. Al final, cada pareja debe compartir con la clase un ejemplo de lo que discutieron.

---

### **3. Retroalimentación y Discusión (5 minutos)**  
- **Discusión:** Invitar a algunos estudiantes a compartir sus experiencias de la actividad. 
- **Preguntas:** ¿Cómo se sintieron al intentar ser asertivos? ¿Les resultó difícil? ¿Cómo fue recibido el mensaje?
  
**Consejo:** Resaltar la importancia de practicar la asertividad como una herramienta para mejorar la comunicación interpersonal y reducir conflictos.

---

### **Cierre**  
- Resumir las características de la comunicación asertiva: uso de "yo" en lugar de "tú" para evitar culpabilizar, tono de voz calmado, y respeto hacia el interlocutor.
- **Tarea opcional:** Pedirles que practiquen la comunicación asertiva en sus conversaciones durante la semana y escriban una reflexión breve sobre cómo les fue.

**Material adicional:**  
- Video corto o recurso digital sobre la importancia de la comunicación asertiva.
